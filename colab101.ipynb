# Google Colab とは

Python はプログラミング言語です。

Windows、Mac、Linuxなど、色々な環境にインストールしてプログラム開発ができます。

Pythonのプログラム開発を、Webブラウザ上で出来るようにしたのが、Jupyter Notebook（ジュピターノートブック）です。Webサーバーにインストールして動かします。

Jupyter Notebookをベースに、ライブラリなどを加え、すぐに使えるように公開しているサービスが、Google Colaboratry（略称 Colab）です。

Colaboratory よくある質問 (https://research.google.com/colaboratory/faq.html?hl=ja) もみてください。

Jupyter Notebookは、以前は「IPython Notebook」という名前だったため、ファイル拡張子は ipynb です。

| [shift]+[→] | 文字選択 |
| [ ctrl ]+[ c ] | コピー |
