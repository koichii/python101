## ショートカット

文字を選択したり、切り貼りをするには、ショートカットが便利です。
いくつかのショートカットは、Scratchなど他のアプリでも使えます。

| キー操作 | 動作 |
|:--:|:--|
| [shift]+[↑] | 行選択 |
| [shift]+[↓] | 行選択 |
| [shift]+[←] | 文字選択 |
| [shift]+[→] | 文字選択 |
| [ ctrl ]+[ a ] | 全選択 |
| [ ctrl ]+[ x ] | カット（切り取り） |
| [ ctrl ]+[ c ] | コピー |
| [ ctrl ]+[ v ] | ペースト（貼り付け） |
| [ ctrl ]+[ z ] | 取消 |
| [ tab  ] | インデント（字下げ。スペース4つ） |
| [shift]+[ tab ] | インデント解除 |

※『＋』は「押したまま」と言う意味です。例えば [ ctrl ]+[ c ] は、 [ ctrl ]を押したまま（離さずに） [ c ] を押し、すぐに [ c ] を離します。

※ マックでは [ ctrl ] では無く ⌘ キーを使います。

# タイピング ゲーム

表示された単語を正しく入力して、その早さを競います。  

Scratch版タイピング ゲーム：
https://scratch.mit.edu/projects/659945072

Scratch版と、Python版と くらべてみよう。


```python
# Python版タイピング ゲーム
import time

mondai = ["news", "hello", "girl", "chair", "desk"] 
hajime = time.time()
i = 0
while i < len(mondai):
    Q = mondai[i]
    A = input(Q + "といれて：")
    if Q == A:
        print("せいかい！")
        i = i + 1
    else:
        print("まちがい!")
owari = time.time()
print(owari - hajime, "びょうでした")
```
