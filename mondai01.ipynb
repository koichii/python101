# 給料から税金と社会保険料を計算する

- 基礎控除は48万円
- 給与所得

給与年収180万円以下の場合、控除額は給与年収×40%-100,000円となります。
ただし、計算した給与所得控除額が55万円未満のときは55万円です。
180万円超～360万円以下の場合、控除額は給与年収×30%＋8万円となります。
360万円超～660万円以下の場合、控除額は給与年収×20%＋44万円となります。
660万円超～1,000万円以下の場合、控除額は給与年収×10%＋110万円となります。
850万円超を超える場合の控除額は195万円（上限）です。

- 例えば
月22万円×12カ月＝年予算264万円の場合：

消費税10%分差引：
264万 / 1.1 = 240万円（24万円）

- 厚生年金保険料 18.3%
240万円×18.3% = 43.92万
控除金額は半額
43.92 / 2 = 21.96万

給与所得控除：

240万×30%＋8万=80万
給与所得：
240万 - 80万 - 21.96万 = 138.04万
基礎控除：
48万
課税所得：
138.04万 - 48万 = 90.04万

所得税（5.105%）住民税（10%）：
90.04万×15.105% = 13.6万

税金＋社会保険料：81.52万

```python
# 給与所得を計算する
def zeikin(kyuyo):
    return kyuyo

# 給料を入力して、税金を計算する
gekkyu = int(input("月給は："))
nenshu = gekkyu * 12
print("年収：", nenshu)
shotoku = zeikin(nenshu)
print("所得：", shotoku)
print("税金：", shotoku * 0.15105)
```
