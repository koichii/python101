## インスタンスとオブジェクト参照（さんしょう）
作られたオブジェクトの**メモリ上の実体**を、オブジェクトの「インスタンス」と言います。

オブジェクト参照は、メモリ上の実体が「どこにあるか」の情報だけが入っています。

そのため、オブジェクト参照を別の変数に入れても、オブジェクトのインスタンスが増えることはありません。元のオブジェクト参照も、オブジェクト参照のコピーも、同じオブジェクトのインスタンス（メモリ上の実体）を参照します。

## クラス
クラスは、オブジェクトの「ひな形」「テンプレート」「設計図」です。そのオブジェクトには、何と言う変数と関数が定義されているのかを決めています。

下のプログラムでは、クラスを定義して、そのクラスのオブジェクトを2つ（obj1とobj2を）作り、オブジェクトの変数を print しています。

````python
# オブジェクトのクラス定義
class Object1:
    def __init__(self, x=0, y=0):   # オブジェクトの変数を初期化するコンストラクタ。__init__ と言う名前を使う。
        self.x = x
        self.y = y

obj1 = Object1(10,30)       # オブジェクトを作って、参照を obj1 変数に入れる
obj2 = Object1(20,40)       # オブジェクトを作って、参照を obj2 変数に入れる
print(obj1.x, obj1.y)       # obj1のx、obj1のy を print
print(obj2.x, obj2.y)       # obj2のx、obj2のy を print
````

## コンストラクタ
クラスからオブジェクトを作るときに、変数の初期化などを行う、特別のメソッドを「コンストラクタ」といいます。Pythonでは、『 ```__init__``` 』（イニット）と言う名前に決まっています。

クラス定義の中では『self』変数が、インスタンスを参照しています。

````python
# オブジェクトのクラス定義
class Object1:

    def __init__(self, x=0, y=0):   # オブジェクトの変数を初期化するコンストラクタ
        self.x = x
        self.y = y

    def reverse(self):              # 自分の好きな関数（メソッド）を追加できる
        self.x *= -1
        self.y *= -1    

obj1 = Object1(-30,40)      # オブジェクトを作って、参照を bowl 変数に入れる
print(obj1.x, obj1.y)
````

````python
obj1.reverse()              # 自分が作ったメソッドをよびだす
print(obj1.x, obj1.y)
````

オブジェクトの変数は、普通の変数と同じように、値を変えたり、計算したりできます。

````python
obj1.y = 10
print(obj1.y)
obj1.y += 40
print(obj1.y)
````

オブジェクトの変数は、class（クラス）で定義するのが基本ですが、あとから、好きにふやすことができます。

````python
obj1.z = 70    # z を追加する
print(obj1.z)
````


## オブジェクト参照（さんしょう）
オブジェクト参照の obj1 を別の変数、例えば obj3 に入れても、参照をコピーするだけで、インスタンス（オブジェクトの実体）はコピーされません。

obj1 も obj3 も、同じインスタンス（メモリ上の実体）を、参照します。

````python
obj3 = obj1
print(obj3.y, obj1.y)
obj3.y += 5
print(obj3.y, obj1.y)
````

オブジェクトの中にオブジェクトを入れることもできます。

````python
obj1.pos = Object1(10, 20)
print(obj1.pos.x, obj1.pos.y)
````

ちなみに、obj1 と obj3 は、今でも同じインスタンスを参照しているので、どちらからでも pos の中を見る事ができます。

````python
print(obj1.pos.x, obj1.pos.y)
print(obj3.pos.x, obj3.pos.y)
````

オブジェクトをリストに入れることもできます。

````python
obj7 = [Object1(10,20), Object1(30,40), Object1(50,60)]
for obj6 in obj7:
    print(obj6.x, obj6.y)
````

# 実は、すべてがオブジェクト
実は、Pythonは、数値や文字列もふくめ、すべての種類の変数がオブジェクトです。

"文字列".format() は、文字列オブジェクトのメソッドです。

しかしながら、数値や文字列はイミュータブル（変更不可）オブジェクトであるため、オブジェクトっぽくない動きをするようになっています。

イミュータブル（変更不可）オブジェクトは、オブジェクト参照をコピーすると、必要に応じてインスタンス（メモリ上の実体）もコピーされ、別々のインスタンスになります。そのため、関数呼び出しの中などでは、元のオブジェクトの中を変更することができないようになっています。
このため、イミュータブル（変更不可）オブジェクトは、（オブジェクトでは無く）値そのもののように扱うことができます。

一方、リストやclassを使って定義したオブジェクトは、ミュータブル（変更可能）オブジェクトです。オブジェクト参照をコピーしてもインスタンス（メモリ上の実体）は１つのままで、どこからでも、元のオブジェクトの中の変数を変更可能です。
