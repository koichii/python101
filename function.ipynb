# 1 関数（かんすう）とは

関数は、値（あたい）を受け取って、処理をして、値を返します。

````python
# len（レン） 関数は、リストを受け取って、リストの長さ（リストに含まれる項目の個数）を返します。

lst = [1, 2, 3]
a = len(lst)
print(a)

lst = [4, 1, 3, 9]
a = len(lst)
print(a)

lst = [4, 1, 3, 9, 2]
print( len(lst) )
````

````python
# sum （サム）関数は、リストを受け取って、リストの中の項目をすべて合計した結果を返します。

lst = [1, 2, 3]
a = sum(lst)
print(a)

lst = [4, 1, 3, 9]
a = sum(lst)
print(a)

lst = [4, 1, 3, 9, 2]
print( sum(lst) )
````

````python
# sum 関数と len 関数を使って、平均を計算することもできます。

lst = [1, 2, 3]
a = sum(lst) / len(lst)
print(a)

lst = [4, 1, 3, 9]
a = sum(lst) / len(lst)
print(a)

lst = [4, 1, 3, 9, 2]
print( sum(lst) / len(lst) )
````

## 1-2 ユーザー定義関数（ていぎかんすう）
もともと用意されていなくて、プログラマーが後から作った関数を、ユーザー定義関数といいます。

関数を作ることを、関数を定義（ていぎ）する、といいます。『定義する』を、英語で define（デファイン）といいます。

関数の名前を、関数名といいます。  
関数が受け取る値を、引数（ひきすう）といいます。  
関数が返す値を、戻り値（もどりち）といいます。
```
# 関数の定義

def 関数名(引数1, 引数2, ...):
    処理1
    処理2
    return 戻り値
```

````python
# add 関数を定義する（引数 a と b を足し算した答えが戻り値）
def add(p1, p2):
    a = p1 + p2
    return a

# add 関数を呼び出す
print("足し算：", add(3, 4) )
print("足し算：", add(-5, -2) )
print("足し算：", add(31, 5) )
````

## 2-2 文字列を操作する関数をつくろう

文字列は、リストと同じように、[ ] を使って、1文字を取り出すことができます。

また、 for ～ in の繰り返しを使って、1文字づつ文字を取り出すこともできます。

```
for 変数 in 文字列:    # 変数には、繰り返しの度に、文字列の中の1文字が入る
    処理1
    処理2
    ...
```

````python
# 文字列の中の文字を探す
MOJI = "ABCD"

i = 0
for j in MOJI:
    print(i, "番目は", MOJI[i],"そして", j)
    if j == MOJI[2]:
        print("MOJI[2]である")
    else:
        print("MOJI[2]では無い")
    print("")
    i += 1
````
