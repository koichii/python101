# 関数（かんすう）

関数は、値（あたい）を受け取って、処理をして、値を返します。  
受け取る値を、引数（ひきすう）といいます。  
返す値を、戻り値（もどりち）といいます。

#組み込み関数

あらかじめ用意されている関数を組み込み関数と言います。Python には、69個の組み込み関数があります。

python 組み込み関数を全て(69個)紹介する  
https://qiita.com/anieca/items/5afa1787aa0c7cd3ab49

````python
# 組み込み関数
# print関数は、引数を、プリントします。
# max 関数は、引数としてリストを受け取り、戻り値として最大値を返します。
# min 関数は、引数としてリストを受け取り、戻り値として最小値を返します。
# sum 関数は、引数としてリストを受け取り、戻り値として合計を返します。
# len 関数は、引数としてリストを受け取り、戻り値としてリストの長さ（個数）を返します。

lst = [4, 1, 3, 9]
print("リスト:", lst)
print("最大 max:", max(lst))
print("最小 min:", min(lst))
print("合計 sum:", sum(lst))
print("長さ len:", len(lst))
print("平均 average:", sum(lst)/len(lst))

lst = [5, 8, 2, -7, 3]
print("リスト:", lst)
print("最大 max:", max(lst))
print("最小 min:", min(lst))
print("合計 sum:", sum(lst))
print("長さ len:", len(lst))
print("平均 average:", sum(lst)/len(lst))
````

## ユーザー定義関数（ていぎかんすう）
もともと用意されていなくて、プログラマーが後から作った関数を、ユーザー定義関数といいます。

関数を作ることを、定義（ていぎ）する、といいます。

```
# 関数の定義

def 関数名(引数1, 引数2, ...):
    処理1
    処理2
    return 戻り値
```



````python
# 引数 a と b を受け取って、足し算した答えを返す関数
def plus(a, b):
    c = a + b
    return c

print("3 + 4 =", plus(3, 4))
````
