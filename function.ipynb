# 関数（かんすう）

関数は、値（あたい）を受け取って、処理をして、値を返します。

````python
# len（レン） 関数は、リストを受け取って、リストの長さ（リストに含まれる項目の個数）を返します。

lst = [1, 2, 3]
a = len(lst)
print(a)

lst = [4, 1, 3, 9]
a = len(lst)
print(a)

lst = [4, 1, 3, 9, 2]
print( len(lst) )
````

````python
# sum （サム）関数は、リストを受け取って、リストの中の項目をすべて合計した結果を返します。

lst = [1, 2, 3]
a = sum(lst)
print(a)

lst = [4, 1, 3, 9]
a = sum(lst)
print(a)

lst = [4, 1, 3, 9, 2]
print( sum(lst) )
````

````python
# sum 関数と len 関数を使って、平均を計算することもできます。

lst = [1, 2, 3]
a = sum(lst) / len(lst)
print(a)

lst = [4, 1, 3, 9]
a = sum(lst) / len(lst)
print(a)

lst = [4, 1, 3, 9, 2]
print( sum(lst) / len(lst) )
````

#組み込み関数

あらかじめ用意されている関数を組み込み関数と言います。Python には、69個の組み込み関数があります。

python 組み込み関数を全て(69個)紹介する  
https://qiita.com/anieca/items/5afa1787aa0c7cd3ab49

そのうちの、6個の組み込み関数を説明します。

````python
# 組み込み関数
# print（プリント）関数は、引数を、プリントします。
# max（マックス）関数は、リストを受け取り、最大値を返します。
# min（ミン）関数は、リストを受け取り、最小値を返します。
# sum（サム）関数は、リストを受け取り、合計を返します。
# len（レン）関数は、リストを受け取り、リストの長さ（項目数）を返します。
# abs（アブス）関数は、値を受け取り、絶対値を返します。

lst = [4, 1, 3, 9]
print("リスト:", lst)
print("最大 max:", max(lst) )
print("最小 min:", min(lst) )
print("合計 sum:", sum(lst) )
print("長さ len:", len(lst) )
print("-3の絶対値 abs", abs(-3) )

lst = [5, 8, 2, -7, 3]
print("リスト:", lst)
print("最大 max:", max(lst) )
print("最小 min:", min(lst) )
print("合計 sum:", sum(lst) )
print("長さ len:", len(lst) )
print("lst[3]の絶対値 abs", abs(lst[3]) )
````

## ユーザー定義関数（ていぎかんすう）
もともと用意されていなくて、プログラマーが後から作った関数を、ユーザー定義関数といいます。

関数を作ることを、関数を定義（ていぎ）する、といいます。『定義する』を、英語で define（デファイン）といいます。

関数の名前を、関数名といいます。  
関数が受け取る値を、引数（ひきすう）といいます。  
関数が返す値を、戻り値（もどりち）といいます。
```
# 関数の定義

def 関数名(引数1, 引数2, ...):
    処理1
    処理2
    return 戻り値
```

````python
# add 関数を定義する（引数 a と b を足し算した答えが戻り値）
def add(a, b):
    c = a + b
    return c

# add 関数を呼び出す
print("3 + 4 =", add(3, 4) )
````
