```python
import nest_asyncio
nest_asyncio.apply()
import asyncio
import websockets
import json

async def recive(ws):
    while True:
        print("受信")

async def main(uri):
    async with websockets.connect(uri) as websocket:
        # task1 = asyncio.create_task(recive(websocket))
        await websocket.send(json.dumps({
            "MSGTYPE": "JOIN",
            "room": "python101@j-code.org",
        }))
        while True:
            print(await websocket.recv())
            msg = input()
            await websocket.send(json.dumps({
                "MSGTYPE": "MESSAGE",
                "sendto": "*",
                "chat": msg,
            }))

print("AA")
asyncio.run(main("wss://j-code.org/ws/"))
print("BB")
```

## 関数（かんすう）
手順をまとめて一発で呼び出せるようにしたものを関数（かんすう）、あるいはファンクションと言います。
関数に受け渡すパラメータを「ひきすう」といいます。
関数から返す値を「もどりち」といいます。

```python
# 関数（かんすう）ファンクション
# 「〇〇といれて」と 表示して入力をまつ
def input2(hikisu):
    print(hikisu, "といれて")
    a = input()
    return a
```
