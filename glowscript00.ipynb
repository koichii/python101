# Glowscript（グロウスクリプト）VPython（ブイパイソン） とは

VPythonは、Pythonプログラミングと「Visual（ビジュアル）」と呼ばれる3Dグラフィックモジュールをくみあわせたものです。

VPythonを使用すると、ユーザーは球体（きゅうたい）や円錐（えんすい）などのオブジェクトを3Dで作成し、表示できます。
この VPythonを Web上で動かすようにしたものが Glowscript( https://glowscript.org/ ) です。

下のがめんの【**▶Run**】をクリックしてください。プログラムが動きます。

```
import IPython
IPython.display.IFrame(src="https://trinket.io/embed/glowscript/1f9fe75f47", width="100%", height="480")
```

カメラを回転、ズーム、パンすることで、見え方がかわります。
- 回転: [Ctrl]+マウス左ドラッグ、または、マウス右ドラッグ
- ズーム: [Alt]+マウス左ドラッグ、または、マウスのホイール 
- パン: [Shift]＋マウス左ドラッグ 

# trinket.io とは

trinket.io では、Glowscript VPython プログラムを作ることができます。
trinket.io にアカウントを作成すると、自分が作ったプログラムを保存できます。
日本語版はありませんが、使いやすいので、すぐに使い方をおぼえることができると思います。

※アカウントを作成しなくても、プログラムを動かすことはできます。

## サインアップ

- https://trinket.io/ をひらく

- 下の画面があらわれたら、【**Sign Up**】をクリック
