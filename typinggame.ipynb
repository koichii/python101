# タイピング ゲーム

表示された単語を正しく入力して、その早さを競います

```python
# タイピング ゲーム
import time

mondai = ["news", "hello", "girl", "chair", "desk"] 
hajime = time.time()
i = 0
while i < len(mondai):
    Q = mondai[i]
    print(Q, " と いれて")
    A = input()
    if Q == A:
        print("せいかい！")
        i = i + 1
    else:
        print("まちがい!")
owari = time.time()
print(owari - hajime, "びょうでした")
```
